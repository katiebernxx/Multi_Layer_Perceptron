```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import load_stl10_dataset
import preprocess_data

plt.show()
plt.style.use(['seaborn-v0_8-colorblind', 'seaborn-v0_8-darkgrid'])
plt.rcParams.update({'font.size': 20})

np.set_printoptions(suppress=True, precision=3)

%load_ext autoreload
%autoreload 2
```

```python
stl_imgs, stl_labels = load_stl10_dataset.load()
```

```python
classes = np.loadtxt(os.path.join('data', 'stl10_binary', 'class_names.txt'), dtype=str)
```

```python
#randomly select 25 images 
indices = random.sample(range(len(stl_imgs)), 25)
selected_images = stl_imgs[indices]
selected_labels = stl_labels[indices]

#5x5 plot
fig, axes = plt.subplots(5, 5, figsize=(8, 8))

for i, ax in enumerate(axes.flat):
    ax.imshow(selected_images[i])
    ax.set_title(f'Class: {selected_labels[i]}', size = 15)
    ax.axis('off') 

plt.tight_layout()
plt.show()
```
